```python
mnist = datasets.MNIST("/home/taanish/MNIST", train=True, download=False)
```

```python
X = mnist.data.numpy()
X = np.reshape(X, (60000, 784))
labels = mnist.targets.numpy()[:60000]
```

```python
k = knn.KNN(X, labels)
```

```python
k_scs = knn.vary_knn(n=10000, m=1000, trials=3, k=k, steps=[10,20,50,100])
```

```python
k_scs_np = np.array(k_scs)
```

```python
plt.plot([10,20,50,100], [np.mean([k_scs_np[i][j] for i in range(3)]) for j in range(4)])
```

```python
k_values = [10, 20, 50, 100]
orig_values = []
oose_values = []
```

```python
k.generate_knns(10000, 1000)
```

```python
for k_val in k_values:
    orig, oose = knn.compare_distributions(k, k_val)
    orig_values.append(orig)
    oose_values.append(oose)
```

```python
oose_values
```

```python
plt.plot(k_values, orig_values, label="Original")
plt.plot(k_values, oose_values, label="OOSE")
plt.show()
```
